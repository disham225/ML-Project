# **Project Summary -**

Project Summary: Predicting Email Campaign Effectiveness Using Machine Learning
In the ever-evolving digital marketing landscape, email campaigns play a pivotal role in customer engagement, retention, and conversion. However, the effectiveness of these campaigns is often challenged by varying customer responses, necessitating a data-driven approach to optimize outcomes. This project aimed to predict the effectiveness of email campaigns using machine learning techniques, offering valuable insights that can enhance campaign strategies, improve customer targeting, and ultimately drive business growth.

Objective

The primary objective of this project was to build a predictive model capable of determining the likelihood of an email being successful, as defined by customer engagement metrics. Success was measured by the Email_Status variable, which indicated whether a customer interacted positively with the email. By leveraging machine learning, the goal was to identify patterns in the dataset that could predict this outcome, allowing for more informed decision-making in future campaigns.

Dataset Overview

The dataset comprised 68,353 rows and 12 primary columns, each representing various attributes related to the email campaigns and customer characteristics. Key features included Subject_Hotness_Score, Total_Past_Communications, Time_Email_sent_Category, Word_Count, Total_Links, Total_Images, and Email_Status. Additionally, the dataset included binary-encoded categorical variables such as Customer_Location_B, Customer_Location_C, and others, along with Email_Length_Category and Email_Type.

The dataset posed challenges, including missing values in key columns like Customer_Location, Total_Past_Communications, Total_Links, and Total_Images. These missing values were imputed using statistical methods, ensuring the dataset was robust enough for model training.

Methodology

The project employed three machine learning models: Logistic Regression, Gradient Boosting Classifier, and Support Vector Machine (SVM). These models were chosen for their varying strengths in classification tasks, enabling a comparative analysis to identify the most effective approach.

Model 1: Logistic Regression
Logistic Regression was used as the baseline model due to its simplicity and interpretability. It provided a clear understanding of how each feature contributed to the prediction of email effectiveness. The model was evaluated using standard metrics such as accuracy, precision, recall, F1 score, and ROC AUC. The initial results were promising, with balanced performance across all metrics.

Model 2: Gradient Boosting Classifier
The Gradient Boosting Classifier was selected for its ability to handle complex, non-linear relationships in the data. This model showed improved performance over Logistic Regression, particularly in recall and ROC AUC, which are critical in identifying successful email campaigns. Cross-validation and hyperparameter tuning were performed using GridSearchCV, leading to further improvements in model accuracy and robustness.

Model 3: Support Vector Machine (SVM)
SVM was used for its strength in high-dimensional spaces and its effectiveness in scenarios where classes are not linearly separable. Despite its theoretical advantages, the model's training time was significantly longer, especially during hyperparameter tuning with RandomizedSearchCV. Ultimately, the performance gains did not justify the computational cost, making it less practical for large-scale application.

Results and Insights

After thorough evaluation, the Gradient Boosting Classifier emerged as the most suitable model for this task. It provided the best balance between precision and recall, essential for minimizing false positives and negatives in email campaign predictions. The ROC AUC score further confirmed its ability to distinguish between successful and unsuccessful campaigns effectively.

To understand the business implications of these models, SHAP (SHapley Additive exPlanations) was initially considered for feature importance analysis. However, due to compatibility issues with multiclass classification in Gradient Boosting, alternative tools like Eli5 were explored. Eli5 provided clear visualizations of feature importance, highlighting key drivers such as Subject_Hotness_Score and Total_Past_Communications.

Conclusion

This project demonstrated the power of machine learning in predicting email campaign effectiveness, offering a data-driven approach to optimize marketing strategies. By selecting the Gradient Boosting Classifier as the final model, the project provided a robust tool for improving customer engagement and maximizing the return on investment for email campaigns. The insights gained from feature importance analysis can guide future campaign designs, ensuring they resonate more effectively with the target audience. Moving forward, integrating this model into a real-time marketing system could further enhance its impact, enabling dynamic, responsive campaign adjustments based on predictive analytics.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Display the first few rows
print("Initial Data:")
print(df.head())

# Display dataset information
print("\nData Info:")
print(df.info())

# Check for duplicates
print("\nDuplicate Rows:")
print(f'Duplicate rows: {df.duplicated().sum()}')

# Check for missing values
print("\nMissing Values Info:")
print(df.isnull().sum())

````

````python
# Handle missing values for numerical columns
numerical_cols = ['Total_Past_Communications', 'Total_Links', 'Total_Images']
numerical_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = numerical_imputer.fit_transform(df[numerical_cols])

# Handle missing values for categorical columns
categorical_cols = ['Customer_Location']
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])



````

````python
# Convert categorical variables to 'category' dtype
df['Email_Type'] = df['Email_Type'].astype('category')
df['Email_Source_Type'] = df['Email_Source_Type'].astype('category')
df['Email_Campaign_Type'] = df['Email_Campaign_Type'].astype('category')
df['Time_Email_sent_Category'] = df['Time_Email_sent_Category'].astype('category')
````

````python
# Create a new feature for email length category
df['Email_Length_Category'] = pd.cut(df['Word_Count'], bins=[0, 100, 500, 1000, 5000], labels=['Short', 'Medium', 'Long', 'Very Long'])

# Convert 'Email_Length_Category' to a categorical variable
df['Email_Length_Category'] = df['Email_Length_Category'].astype('category')

# Encode categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=['Customer_Location', 'Email_Length_Category'], drop_first=True)

# drop Email column
df.drop('Email_ID', axis=1, inplace=True)


# Define feature columns and target variable
X = df.drop('Email_Status', axis=1)
y = df['Email_Status']

# Display the shapes of X and y
print("\nFeature and Target Shapes:")
print(f'Features shape: {X.shape}')
print(f'Target shape: {y.shape}')

# Display the first few rows of the processed data
print("\nProcessed Data:")
print(X.head())
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* Null Hypothesis (H0): There is no significant difference in Subject_Hotness_Score between emails sent at different times of the day (i.e., Time_Email_sent_Category).
* Alternate Hypothesis (H1): There is a significant difference in Subject_Hotness_Score between emails sent at different times of the day.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Group by 'Time_Email_sent_Category' and collect 'Subject_Hotness_Score'
grouped_scores = [group['Subject_Hotness_Score'] for name, group in df.groupby('Time_Email_sent_Category')]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*grouped_scores)
print(f'F-Statistic: {f_stat}, P-Value: {p_value}')
````

##### Which statistical test have you done to obtain P-Value?

* We performed an ANOVA (Analysis of Variance) test to compare the means of Subject_Hotness_Score across different categories of Time_Email_sent_Category.

##### Why did you choose the specific statistical test?

* ANOVA is used when comparing means across multiple groups. Here, it helps determine if there is a significant difference in Subject_Hotness_Score across different times of the day.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* Null Hypothesis (H0): The Total_Links and Total_Images are independent of each other (i.e., there is no significant correlation between them).
* Alternate Hypothesis (H1): The Total_Links and Total_Images are not independent of each other (i.e., there is a significant correlation between them).

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Perform Pearson correlation test
correlation, p_value = stats.pearsonr(df['Total_Links'].dropna(), df['Total_Images'].dropna())
print(f'Correlation: {correlation}, P-Value: {p_value}')
````

##### Which statistical test have you done to obtain P-Value?

* We performed a Pearson correlation test to assess the linear relationship between Total_Links and Total_Images

##### Why did you choose the specific statistical test?

* Pearson correlation is appropriate for evaluating the strength and direction of a linear relationship between two continuous variables.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* Null Hypothesis (H0): The average Word_Count is the same for emails with Email_Status 0 and Email_Status 1.
* Alternate Hypothesis (H1): The average Word_Count is different for emails with Email_Status 0 and Email_Status 1.

#### 2. Perform an appropriate statistical test.

````python
# Extracting the Word_Count for Email_Status 0 and 1
word_count_status_0 = df[df['Email_Status'] == 0]['Word_Count']
word_count_status_1 = df[df['Email_Status'] == 1]['Word_Count']

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(word_count_status_0, word_count_status_1, equal_var=False)
print(f'T-Statistic: {t_stat}, P-Value: {p_value}')
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

def cap_outliers(df, column):
    upper_limit = df[column].mean() + 3*df[column].std()
    lower_limit = df[column].mean() - 3*df[column].std()
    df[column] = np.where(df[column] > upper_limit, upper_limit, df[column])
    df[column] = np.where(df[column] < lower_limit, lower_limit, df[column])

cap_outliers(df, 'Total_Past_Communications')
cap_outliers(df, 'Total_Links')
cap_outliers(df, 'Total_Images')

print(df.describe())
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Such imbalances can lead to several challenges:

Model Bias: Machine learning models may become biased towards the majority class, leading to skewed performance where the model predominantly predicts the majority class. This can adversely affect the predictive accuracy for the minority classes.

Evaluation Metrics: Standard accuracy can be misleading in imbalanced datasets. Metrics such as precision, recall, F1-score, and area under the ROC curve (AUC-ROC) are more informative for assessing model performance on minority classes.

Learning Challenges: The model may struggle to capture the patterns associated with the minority classes due to their limited representation, potentially leading to poor generalization.

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

* To address the imbalance in the dataset, I employed the SMOTE (Synthetic Minority Over-sampling Technique) technique. SMOTE is a popular method for handling imbalanced datasets by generating synthetic samples for the minority class.

Reason for Using SMOTE:
Balancing Class Distribution: SMOTE works by creating synthetic samples in the feature space for the minority class. This process balances the class distribution by increasing the number of samples in the underrepresented classes, which helps in mitigating the bias towards the majority class.

Improving Model Performance: By providing a more balanced dataset, SMOTE helps the model learn better patterns for the minority classes, potentially leading to improved performance metrics such as precision, recall, and F1-score for those classes.

Avoiding Overfitting: Unlike simple oversampling methods that duplicate existing samples, SMOTE generates new, unique samples. This reduces the risk of overfitting that can occur with repeated instances of the same data points.

Enhancing Generalization: With a more balanced training set, the model is better equipped to generalize across all classes, resulting in more reliable predictions and a better understanding of the minority class characteristics.

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

I considered the following evaluation metrics for a positive business impact:

Accuracy: It provides a general overview of the model's performance by showing the proportion of correctly predicted instances. High accuracy indicates that the model is making correct predictions most of the time, which is crucial for maintaining the credibility of the email campaigns.

Precision: Precision is important as it measures the accuracy of positive predictions. In the context of email campaigns, high precision ensures that when the model predicts a successful campaign (positive class), it is likely to be correct. This minimizes the risk of wasting resources on campaigns that are predicted to be successful but actually are not.

Recall: Recall is crucial for identifying actual successful campaigns (true positives). High recall ensures that most of the successful campaigns are identified, reducing the chance of missing out on potential successful email strategies.

F1 Score: The F1 Score is the harmonic mean of precision and recall, providing a balanced view of the model's performance. It is particularly useful when dealing with imbalanced datasets, ensuring that both false positives and false negatives are minimized.

ROC AUC: ROC AUC (Receiver Operating Characteristic Area Under Curve) measures the model's ability to distinguish between classes. A high ROC AUC value indicates that the model is good at distinguishing between successful and unsuccessful campaigns. This is important for making informed decisions on which campaigns to focus on.

Working: The model starts with an initial prediction and iteratively adds trees to the model. Each new tree is fit on the residual errors of the previous ensemble of trees. The process continues until the desired number of trees is added or the performance improvement becomes negligible.
